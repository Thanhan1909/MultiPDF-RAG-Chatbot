```python
import os
import shutil
import zipfile
from pathlib import Path

# --- Cбә¬P NHбә¬T ДҗЖҜб»ңNG IMPORT THEO CHUбәЁN Mб»ҡI ---
from langchain_community.document_loaders import PyPDFLoader  # Vбә«n giб»Ҝ nguyГӘn do pypdf thuб»ҷc loбәЎi loader cЖЎ bбәЈn
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings      # Thay cho langchain_community.embeddings
from langchain_chroma import Chroma                         # ThЖ° viб»Үn Д‘б»ҷc lбәӯp mб»ӣi cho Chroma
import torch
from IPython.display import FileLink
```

```python


# --- CбәӨU HГҢNH ДҗЖҜб»ңNG DбәӘN ДҗбәҰU VГҖO TRб»°C TIбәҫP ---
DATA_DIR = "/kaggle/input/datasets/anhoa1/ragpdf/arxiv_papers/papers"  # Дҗб»Қc trб»ұc tiбәҝp tб»« thЖ° mб»Ҙc sбәЎch vб»«a crawl б»ҹ Гҙ trЖ°б»ӣc
DB_DIR = "./chroma_db"
ZIP_OUTPUT_PATH = "./chroma_db_output.zip"

print("рҹ“Ӯ 1 & 2 & 3. QuГ©t thЖ° mб»Ҙc dб»Ҝ liб»Үu bГ i bГЎo vб»«a crawl...")
if not os.path.exists(DATA_DIR):
    raise FileNotFoundError(f"вқҢ KhГҙng tГ¬m thбәҘy thЖ° mб»Ҙc '{DATA_DIR}'. HГЈy chбәҜc chбәҜn bбәЎn Д‘ГЈ chбәЎy Гҙ code crawl dб»Ҝ liб»Үu б»ҹ phГӯa trГӘn!")

# TГ¬m tбәҘt cбәЈ cГЎc file PDF trong thЖ° mб»Ҙc papers vГ  cГЎc thЖ° mб»Ҙc con cб»§a nГі
pdf_files = [f for f in Path(DATA_DIR).rglob("*") if f.is_file() and f.suffix.lower() == '.pdf']
print(f"рҹ“„ TГ¬m thбәҘy tб»•ng cб»ҷng {len(pdf_files)} file PDF hб»Јp lб»Ү.")


print("\nрҹ“‘ 4 & 5. Дҗang nбәЎp vДғn bбәЈn vГ  tбәЎo Metadata chuбә©n hГіa...")
raw_docs = []
for fp in pdf_files:
    try:
        # LбәҘy tГӘn chб»§ Д‘б»Ғ (topic) dб»ұa trГӘn tГӘn thЖ° mб»Ҙc cha (vГӯ dб»Ҙ: machine_learning)
        topic = fp.parent.name.replace("_", " ")
        # LбәҘy tГӘn file lГ m title (loбәЎi bб»Ҹ ID arxiv б»ҹ Д‘бә§u nбәҝu cГі dбәЎng 2606.xxxx_Title)
        title = fp.stem.split("_", 1)[-1] if "_" in fp.stem else fp.stem
        
        loader = PyPDFLoader(str(fp))
        for p in loader.load():
            # Bб»• sung siГӘu dб»Ҝ liб»Үu chuбә©n hГіa theo yГӘu cбә§u cб»§a bбәЎn
            p.metadata = {
                "source": str(fp), 
                "page": int(p.metadata.get("page", 0)) + 1, # ДҗЖ°a vб»Ғ Д‘б»Ӣnh dбәЎng trang bбәҜt Д‘бә§u tб»« 1
                "title": title, 
                "topic": topic
            }
            raw_docs.append(p)
    except Exception as e:
        print(f"вҡ пёҸ Bб»Ҹ qua file lб»—i khi Д‘б»Қc: {fp.name} - {e}")

print(f"рҹ“Ҡ Tб»•ng sб»‘ trang tГ i liб»Үu thГҙ thu thбәӯp Д‘Ж°б»Јc: {len(raw_docs)}")


print("\nвңӮпёҸ 6 & 7 & 8. Tiбәҝn hГ nh lб»Қc dб»Ҝ liб»Үu trб»‘ng vГ  phГўn mбәЈnh (Chunking)...")
# LoбәЎi bб»Ҹ cГЎc trang trб»‘ng hoбә·c lб»—i font khГҙng cГі chб»Ҝ
valid_docs = [d for d in raw_docs if d.page_content and d.page_content.strip()]

text_splitter = RecursiveCharacterTextSplitter(chunk_size=800, chunk_overlap=150)
chunks = text_splitter.split_documents(valid_docs)
print(f"рҹ“Ҳ Tб»•ng sб»‘ lЖ°б»Јng chunks Д‘Ж°б»Јc tбәЎo ra thГ nh cГҙng: {len(chunks)}")


```

```python
print("\nвҸі 9 & 10. Khб»ҹi tбәЎo mГҙ hГ¬nh nhГәng BGE-M3 trГӘn GPU/CPU...")
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"рҹ–ҘпёҸ Thiбәҝt bб»Ӣ tГӯnh toГЎn tДғng tб»‘c: {device.upper()}")

embed_model = HuggingFaceEmbeddings(
    model_name="BAAI/bge-m3", 
    model_kwargs={'device': device}, 
    encode_kwargs={'normalize_embeddings': True}
)


print("\nрҹ’ҫ 11 & 12. XГўy dб»ұng vГ  Д‘б»“ng bб»ҷ hГіa cЖЎ sб»ҹ dб»Ҝ liб»Үu ChromaDB...")
if os.path.exists(DB_DIR): 
    shutil.rmtree(DB_DIR)
    
vectordb = Chroma.from_documents(documents=chunks, embedding=embed_model, persist_directory=DB_DIR)
print(f"вң… ДҗГЈ lЖ°u trб»Ҝ dб»Ҝ liб»Үu ChromaDB thГ nh cГҙng!")


print("\nрҹ”Қ 13. Thб»ӯ nghiб»Үm tГӯnh nДғng truy xuбәҘt dб»Ҝ liб»Үu mбә«u (Retrieval)...")
retriever = vectordb.as_retriever(search_kwargs={'k': 3})
res = retriever.invoke("machine learning security analysis")

for i, doc in enumerate(res):
    print(f"  рҹ“Қ Kбәҝt quбәЈ {i+1} | Chб»§ Д‘б»Ғ: [{doc.metadata.get('topic')}] | BГ i viбәҝt: {doc.metadata.get('title')[:40]}... (Trang {doc.metadata.get('page')})")


print("\nрҹ“Ұ 14 -> 17. ДҗГіng gГіi nГ©n sбәЈn phбә©m ChromaDB Д‘б»ғ tбәЈi vб»Ғ...")
with zipfile.ZipFile(ZIP_OUTPUT_PATH, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for root, dirs, files in os.walk(DB_DIR):
        for file in files:
            fp = os.path.join(root, file)
            # Giб»Ҝ nguyГӘn cбәҘu trГәc thЖ° mб»Ҙc nб»ҷi bб»ҷ cб»§a ChromaDB bГӘn trong file zip
            zipf.write(fp, os.path.relpath(fp, os.path.dirname(DB_DIR)))

print(f"вң… XГЎc minh: ДҗГЈ tбәЎo file nГ©n thГ nh phбә©m thГ nh cГҙng tбәЎi {ZIP_OUTPUT_PATH} ({os.path.getsize(ZIP_OUTPUT_PATH)/(1024*1024):.2f} MB)")
print("\nрҹҡҖ HOГҖN THГҖNH TOГҖN Bб»ҳ PIPELINE Sбә CH!")
print("рҹ‘Ү BбәҘm vГ o Д‘Ж°б»қng link bГӘn dЖ°б»ӣi Д‘б»ғ tбәЈi cЖЎ sб»ҹ dб»Ҝ liб»Үu ChromaDB vб»Ғ mГЎy tГӯnh cГЎ nhГўn cб»§a bбәЎn:")
display(FileLink(ZIP_OUTPUT_PATH))
```
